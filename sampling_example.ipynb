# Generating random sampled data

```python
import matplotlib.pyplot as plt
import numpy  as np

import httpimport
with httpimport.remote_repo(['periodic_sampling'], 'https://raw.githubusercontent.com/juanjq/periodic_sampling/main'):
    import periodic_sampling as sampling
        

#generating random data
N = 20
y = np.random.rand(N)
x = np.sort(np.random.rand(N))*N

fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,5))
#plotting the original data
ax1.plot(x,y,color='darkblue')
ax1.set_title('Original data')

#plotting the data sampled periodically
ax2.plot(y,color='deeppink')
ax2.set_title('Periodic sampled data')
plt.show()
```

# Sampling the data

```python
dt=0.5

sample_x,sample_y=sampling.resample(x,y,0.5)
```

# Plotting it

```python
fig, ax = plt.subplots(figsize=(13,9))
ax1=plt.subplot(2,2,1)
ax1.plot(x,y,'-',color='darkblue')
ax1.set_title('Original data')

ax2=plt.subplot(2,2,2)
ax2.plot(y,'-',color='deeppink')
ax2.set_title('Periodic sampled data')


ax3 = fig.add_subplot(fig.add_gridspec(2,2)[1, :])
ax3.plot(x,y,'o-',color='b',label='Original')
ax3.plot(sample_x,sample_y,'.--',color='r',label='Sampled',markersize=9)

ax3.set_title('Re-sampled periodic data, dt = '+str(round(dt,2)))
plt.legend()
plt.show()
```
